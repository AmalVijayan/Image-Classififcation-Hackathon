### Initializing The CNN

```

from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras.layers import Convolution2D
from keras.layers import MaxPooling2D
from keras.layers import Flatten
import pandas as pd


classifier = Sequential()


classifier.add(Convolution2D(filters = 128,kernel_size = (3,3),activation = 'relu', input_shape = (64,64,3)))    #
classifier.add(MaxPooling2D(pool_size = (3,3)))  #


# second Convolutional layer
classifier.add(Convolution2D(64,(3,3),activation = 'relu'))
classifier.add(MaxPooling2D(pool_size = (2,2)))


classifier.add(Flatten())


classifier.add(Dense(units = 500, activation = 'relu'))   #
Dropout(0.5)
classifier.add(Dense(units = 300, activation = 'relu'))   #
Dropout(0.2)
classifier.add(Dense(units = 100, activation = 'relu'))   #
Dropout(0.1)
classifier.add(Dense(units = 50, activation = 'relu'))   #
Dropout(0.1)
#classifier.add(Dense(units = 64, activation = 'relu'))
#Dropout(0.3)

classifier.add(Dense(units = 6, activation = 'softmax'))


```

```
classifier.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['categorical_accuracy'])
```

###Training The CNN

```


from keras.preprocessing.image import ImageDataGenerator

train_dataGen = ImageDataGenerator(rescale = 1./255,
                                   shear_range = 0.2,
                                   zoom_range = 0.2,
                                   horizontal_flip = True)

training_set = train_dataGen.flow_from_directory('gdrive/My Drive/Convolutional_Neural_Network/Intel_challenge/Final_Data/training_set',
                                                target_size = (64,64),
                                                batch_size = 200,
                                                class_mode = 'categorical')
```

```


classifier.fit_generator(training_set,
                        samples_per_epoch = 1000,
                        nb_epoch = 300)



```

### Test Set Predictions

```
data = pd.read_csv("gdrive/My Drive/Convolutional_Neural_Network/Intel_challenge/test_WyRytb0.csv")


```

```
from keras.preprocessing import image


images = data.iloc[:,0].values
img_path = "gdrive/My Drive/Convolutional_Neural_Network/Intel_challenge/Final_Data/test_set/{}"
Y_pred = []
for i in images:
  img = image.load_img(path=img_path.format(i),target_size=(64,64,3))#grayscale=True,
  img = image.img_to_array(img)
  test_img = img.reshape((1,64,64,3))
  img_class = classifier.predict_classes(test_img)
  prediction = img_class[0]
  Y_pred.append(prediction)
  
  
```

```
data['label'] = Y_pred

data.to_csv("gdrive/My Drive/Convolutional_Neural_Network/Intel_challenge/Prediction13.csv", index = False)
```
